```python
#importing 1st data set on the Standard Stats for the team 
StanStats = pd.read_csv('Data/Standard Stats.csv')
StanStats.head()
```

```python
#This will add the latitude and longitude coords for each players birth country
from geopy.exc import GeocoderTimedOut
from geopy.geocoders import Nominatim
   
# declare an empty list to store,latitude and longitude of values & of city column
longitude = []
latitude = []
   
# function to find the coordinate of a given city 
def findGeocode(country):
       
    # try and catch is used to overcome the exception thrown by geolocatorusing geocodertimedout  
    try:   
        #Specify the user_agent as your app name it should not be none
        geolocator = Nominatim(user_agent="your_app_name") 
        return geolocator.geocode(country)
    except GeocoderTimedOut:  
        return findGeocode(country)    
# each value from countryname column will be fetched and sent to function find_geocode   
for i in (StanStats["CountryName"]):
      
    if findGeocode(i) != None:      
        loc = findGeocode(i)
        # coordinates returned from function is stored into two separate lists
        latitude.append(loc.latitude)
        longitude.append(loc.longitude)
    # if coordinate for a city not found, insert "NaN" indicating missing value 
    else:
        latitude.append(np.nan)
        longitude.append(np.nan)
#adding the values as columns into the df
StanStats["Longitude"] = longitude
StanStats["Latitude"] = latitude
#StanStats
```

```python
#Pitbull bc of mr worldwide aka gathering all the players by the country of their origin by their coordinates
Pitbull=StanStats[['Player','CountryName','Latitude','Longitude']]
#P2 will just group all the players by country of birth
Pitbull2 = StanStats.groupby('CountryName', sort=False)['CountryName'].count()

import folium
from folium.plugins import MarkerCluster
#empty map
world_map= folium.Map(tiles="cartodbpositron")
marker_cluster = MarkerCluster().add_to(world_map)
#for each coordinate, create circlemarker of user percent
for i in range(len(StanStats)):
        lat = StanStats.iloc[i]['Latitude']
        long = StanStats.iloc[i]['Longitude']
        radius=5
        popup_text = """Country : {}<br>
                    Number of Players : {}<br>"""
        popup_text = popup_text.format(StanStats.iloc[i]['Country'],
                                   Pitbull2#StanStats.iloc[i]['CountryName']
                                   )
        folium.CircleMarker(location = [lat, long], radius=radius, popup= popup_text, fill =True).add_to(marker_cluster)
        
world_map
```
